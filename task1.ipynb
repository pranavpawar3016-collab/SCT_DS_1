```python
import pandas as pd

data = {
    "Age Group": ["0–20 Years", "21–64 Years", "65+ Years"],
    "Population (Million)": [512, 807, 98]
}

df = pd.DataFrame(data)
df
```

```python


plt.figure(figsize=(8,5))

plt.bar(
    df["Age Group"],
    df["Population (Million)"]
)

plt.xlabel("Age Group")
plt.ylabel("Population (in Millions)")
plt.title("India's Population Distribution by Age Group")

plt.show()
```
